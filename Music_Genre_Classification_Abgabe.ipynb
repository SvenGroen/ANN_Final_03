# Introduction

### Background

In the area of music information retrieval (MIR) different methods are investigated to extract 
useful information from music data. The importance of this research area has been grown with the rise and availability of digital music and music streaming services [1]. Music genre classification is one of the challenges in MIR. [2] explains that music genres are “categorical labels created by humans”. Automating this process is a challenging task, since the labels are based on subjective human categorization, which suffers from vagueness [3]. In this work we will try to replicate the results of [4]. While Huang, Serafini and Pugh investigated multiple approaches towards the problem of genre classification, we will focus on their most successful attempt. The authors have been able to achieve a 82% Accuracy on the training dataset using a Convolutional Neural Network approach with preprocessed data. [4]

### Related Work

Classical approaches towards the music genre classification problem made use of extracting hand crafted features (timbral texture, rhythmic content, etc.) which were used in combination with classical machine learning approaches like k-nearest neighbors and gaussian mixture models to classify the different genres [5]. 

In 2013, [6] pointed out, that progress seemed to plateau in the area of content-based music informatics and elaborated, why hand-crafted features for analysis are suboptimal and why deep learning models achieve potentially better results [6]. Since then, different deep learning approaches have been investigated, such as LSTM-classifier [7], Graph-Neural-Networks[8] or hybrid-architectures [9]. 

Wyse [10] showed that audio data in the form of a spectrogram can be analysed using convolutional neural networks. CNNs for music genre classification are a widely used method[11, 12]. Most of them use the Mel-Spectrogram feature in order to classify audio. 
Mel-Spectrograms project normal spectrograms onto the mel scale. This leads to a visualisation in which low frequency bands are shown very precisely and higher frequency bands are shown in a compressed way [13]. [13] explained that the human ear functions in a similar fashion with higher precision on the lower frequency band. Furthermore, he was able to show that some CNN filters are similar to the e spectrotemporal receptive field from the human auditory system.

# Concept

### Preprocessing

In general, we aim to follow the same preprocessing steps as the authors [1]. 

To do this, we first extract four randomly selected 2-second windows from each audio file. Therefore, we get 400 2-second samples for each genre, that is 4000 samples in total, evenly distributed in terms of genre.

To get the duration of each signal in seconds we need to devide it by the sampling rate: 


$\frac{len(signal)}{sampling\_rate} = duration_{sec.}$

$\Leftrightarrow len(signal) = duration_{sec.} * sampling\_rate$


```python
len_signal = 2 * sampling_rate
print("To receive two-second random clips from our audio signal, we have to look at {} continous entries in our raw numpy signal.".format(len_signal))
```

To achieve this we pick one random starting index and make sure it can maximally be at 28 seconds. 
We then return the signal from that point to the point plus two seconds and do this four times for each audio file.


In the next step, we transform our samples to so-called mel-spectrograms.
Spectrograms are simply-speaking plots visualizing sound. 
Mel-spectrograms are  spectrograms that are "mel-scaled", so that the mel-scale is on its y-axis. 
The mel-scale results from a transformation of the frequency scale, such that it resembles the frequency band that humans can perceive.

Moreover, the authors stated, that they converted the "raw" audio into mel-spectograms, but we are sure they refer to the two-seconds random clips and not the 30-seconds clips. We assume this since the resulting spectogram which you will see below for two seconds looks exactly like the authors' showed in their paper. The "raw" 30 seconds mel-spectogram looks not like the spectograms in the paper, however, we will investigate these later on.

To get the mel-spectogram we use librosas [melspectrogram()](https://librosa.github.io/librosa/generated/librosa.feature.melspectrogram.html) method.

When it comes to the parameters the authors say that they chose the following parameters:
> 64 mel-bins and a window length of 512 samples with an
overlap of 50% between windows. Based on previous academic success with such transformations, we then move to
log-scaling using the formula $log(X^2)$.

We use the exact same parameters when converting our samples to mel-spectrogram.

### Model Architecture

As already mentioned, the model we use is a Convolutional Neural Network (CNN).
As it has many layers and regularizations, we suggest that you take a direct look at our architecture, which you can see in the following code cell.

In general, we try to copy the architecture of [4]. 
However, we notice that using all layers and especially all regularizations, the training tends to take days.
For this reason we decide to leave out some of these, but you can still recognize them as they are commented out.

```python
def cnn(input_shape, num_genres):
    inputs = tf.keras.layers.Input(shape=input_shape)
    
    # 1st Conv-Pool-layers
    x = tf.keras.layers.Conv2D(filters=64, 
                               kernel_size=(4,4), 
                               activation=tf.keras.activations.relu)(inputs)
    x = tf.keras.layers.BatchNormalization()(x)
    x = tf.keras.layers.MaxPool2D(pool_size=(2,4))(x)
    # 2nd Conv-Pool-layers
    x = tf.keras.layers.Conv2D(filters=64, 
                               kernel_size=(3,5), 
                               activation=tf.keras.activations.relu,
                               kernel_regularizer=tf.keras.regularizers.l2(0.04))(x)
    x = tf.keras.layers.MaxPool2D(pool_size=(2,2))(x)
#     x = tf.keras.layers.Dropout(0.2)(x)
    # 3rd Conv-Pool-layers
    x = tf.keras.layers.Conv2D(filters=64, 
                               kernel_size=(2,2), 
                               activation=tf.keras.activations.relu)(x)
    x = tf.keras.layers.BatchNormalization()(x)
    x = tf.keras.layers.MaxPool2D(pool_size=(2,2))(x)
#     x = tf.keras.layers.Dropout(0.2)(x)
    
    # Flatten before feeding into dense layers
    x = tf.keras.layers.Flatten()(x)
    
    # Dense layers
    x = tf.keras.layers.Dense(64,activation=tf.keras.activations.relu, kernel_regularizer=tf.keras.regularizers.l2(0.04))(x)
#     x = tf.keras.layers.Dropout(0.5)(x)
    x = tf.keras.layers.Dense(32,activation=tf.keras.activations.relu, kernel_regularizer=tf.keras.regularizers.l2(0.04))(x)
    
    # output layer
    output = tf.keras.layers.Dense(num_genres, activation=tf.keras.activations.softmax)(x)
    
    cnn_model = tf.keras.Model(inputs=inputs, outputs=output)
    return cnn_model 
```

Concerning the hyperparameters, we use a batchsize of 32, an adam optimizer and a categorical cross-entropy loss.
We train for 50 epochs in total, which takes approximately --- hours.
